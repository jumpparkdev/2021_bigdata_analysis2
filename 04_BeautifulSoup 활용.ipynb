# 4. BeautifulSoup 활용
* 멜론사이트 크롤링하기
* 노래제목, 가수명, 앨범명, 좋아요개수 추출

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.melon.com/chart/index.htm'
headers = {
    'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.131 Safari/537.36'
}
res = requests.get(url, headers = headers)
res
```

```python
res.text[:300]
```

```python
soup = BeautifulSoup(res.text)
```

### 노래 제목 추출

```python
songs = soup.select('.ellipsis.rank01')
songs
```

```python
title = []
for song in songs:
    title.append(song.find('a').get_text())
title
```

### 가수 추출

```python
songs = soup.select('.ellipsis.rank02')
singer = []

for song in songs:
    singer.append(song.find('a').get_text())
singer
```

### 앨범 추출

```python
songs = soup.select('.ellipsis.rank03')
album = []

for song in songs:
    album.append(song.find('a').get_text())
album
```

### 좋아요 추출

```python
## Ajax - Asynchronous JavaScript and XML
## Chrome 기준; 검사 -> Network -> (clear & record network activity) -> Fetch/XHR -> (원하는 거 찾기)
    ## Request URL 및 Query String Headers 등을 보려면 Headers 리본 메뉴 진입
songs = soup.select('.ellipsis.rank03')
singer = []

for song in songs:
    singer.append(song.find('a').get_text())
singer
```

```python
songs = soup.select('.button_etc.like')
params = ''
for song in songs:
    params += song['data-song-no'] + ','
params[:-1]
```

```python
url = 'https://www.melon.com/commonlike/getSongLike.json?contsIds=' + params[:-1]

res = requests.get(url, headers=headers)
res
```

```python
res.json()
```

```python
songs = res.json()['contsLike']
like = []

for song in songs:
    like.append(song['SUMMCNT'])
like
```

```python
# 제목, 가수, 앨범, 좋아요로 dataframe 생성
df_data = {
    'title': title,
    'singer': singer,
    'album': album,
    'like': like
}
df = pd.DataFrame(data=df_data)
df
```
